```python
df = pd.concat([df1, df2], axis = 1, join = 'inner')
df = pd.concat([df, df3], axis = 1, join = 'inner')
df = pd.concat([df, df4], axis = 1, join = 'inner')
df = pd.concat([df, df5], axis = 1, join = 'inner')
```

```python
df.dropna(axis=1, how='all')
df.dropna(axis=0, how='all')
```

```python
df = df.rename(columns = {'SEQN' : 'ID',
                          'RIAGENDR' : 'Gender',
                          'DMDYRSUS' : 'Years_in_US', 
                          'INDFMPIR' : 'Family_income',
                          'LBXGH' : 'GlycoHemoglobin',
                          'BMXARMC' : 'ArmCircum',
                          'BMDAVSAD' : 'SaggitalAbdominal',
                          'MGDCGSZ' : 'GripStrength',
                          'DRABF' : 'Breast_fed'})
```

```python
df = df.loc[:, ['ID', 'Gender', 'Years_in_US', 'Family_income','GlycoHemoglobin', 'ArmCircum', 
                'SaggitalAbdominal', 'GripStrength', 'Breast_fed']]
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dropna(axis=1, how='all')
df.dropna(axis=0, how='all')
```

```python
from sklearn.feature_selection import VarianceThreshold
df['Years_in_US'] = df['Years_in_US'].apply(lambda x: x if x > 0 else 0)

#Populate median values
df['GlycoHemoglobin'] = df['GlycoHemoglobin'].fillna(df['GlycoHemoglobin'].median())
df['SaggitalAbdominal'] = df['SaggitalAbdominal'].fillna(df['SaggitalAbdominal'].median())
df['ArmCircum'] = df['ArmCircum'].fillna(df['ArmCircum'].median())
df['GripStrength'] = df['GripStrength'].fillna(df['GripStrength'].median())

#fill missing values based on last valid observation
df['Family_income'] = df['Family_income'].fillna(method='ffill')

#fill breastfed as 1
df['Breast_fed'] = df['Breast_fed'].fillna(value = 1)

#Remove features that doesnot meet a threshold
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
sel.fit_transform(df)
```

```python
df.loc[df['GlycoHemoglobin'] < 6.0, 'Diabetes'] = 0
df.loc[(df['GlycoHemoglobin'] >= 6.0) & (df['GlycoHemoglobin'] <= 6.4), 'Diabetes'] = 1
df.loc[df['GlycoHemoglobin'] >= 6.5, 'Diabetes'] = 2

df.head(10)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.drop(['GlycoHemoglobin'], axis = 1, inplace=True)

df.head(5)
```

```python
target =  df[['Diabetes']]
att = df.drop(['Diabetes'], axis = 1)
```

```python
#Linear Regression model
from sklearn.linear_model import LinearRegression

#seperate training set and test set
train_X = att[:6000]
test_X = att[6000:]
train_Y = target[:6000]
test_Y = target[6000:]

lr = LinearRegression()
#Train using Linear Regression
lr.fit(train_X, train_Y)

#Predict using test set
lr_predict_diabetes = lr.predict(test_X)
lr_score = lr.score(test_X, test_Y)

print('Linear Regression Coefficients: ', lr.coef_)
print('LR Mean Squared Error: %.2f' % mean_squared_error(test_Y, lr_predict_diabetes))
print('LR Variance score: %.2f' % r2_score(test_Y, lr_predict_diabetes))
print('Score: %.2f' % lr_score)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, tol = 5e-04)
kmeans.fit_predict(train_X)

print('Parameters: ', kmeans.get_params)
print('Predict: ', kmeans.predict)
print('Score: %.2f'% kmeans.score(test_X))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

adboost = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=100, 
                            learning_rate=0.5, random_state=100)
#Train using Adaboost Algorithm
adboost.fit(train_X, train_Y)
print("training:\n")

#PRedict using test data
adboost_predict_diab = adboost.predict(test_X)
print("prediction: \n", adboost_predict_diab)

print("\nparameters: \n", adboost.get_params)

#Score
adboost_score = adboost.score(test_X, test_Y)
print("\nmean accuracy: %.2f" % adboost.score(test_X, test_Y))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier

bagging = BaggingClassifier(base_estimator= DecisionTreeClassifier(), max_samples = 0.5, max_features = 0.5, 
                            bootstrap = False, bootstrap_features = False)
bagging.fit(train_X, train_Y)
bagging_predict_diab = bagging.predict(test_X)
bagging_score = bagging.score(test_X, test_Y)
print("\nmean accuracy(Bagging): %.2f" % bagging.score(test_X, test_Y))


bagging_KN = BaggingClassifier(base_estimator= KNeighborsClassifier(), max_samples = 0.5, max_features = 0.5, 
                            bootstrap = False, bootstrap_features = False)
bagging_KN.fit(train_X, train_Y)
bagging_predict_diab_KN = bagging_KN.predict(test_X)
bagging_KN_score = bagging_KN.score(test_X, test_Y)
print("\nmean accuracy(KNeighbors): %.2f" % bagging_KN.score(test_X, test_Y))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(1000, 300, 300), solver='adam', shuffle=False, tol =1e-4)

mlp.fit(train_X, train_Y)
mlp_predict_diab = mlp.predict(test_X)

print("parameter: ", mlp.get_params())
mlp_score = mlp.score(test_X, test_Y)
print('\nAccuracy: %.2f' % mlp.score(test_X, test_Y))
```

```python
data = {'Model': ['Linear Regression', 'Adaboost', 'Bagging_decision tree based', 'Bagging_KNeighbors', 'MLP'],
     'Accuracy' : [lr_score, adboost_score, bagging_score, bagging_KN_score, mlp_score]}

result_df = pd.DataFrame(data = data)
result_df
```
